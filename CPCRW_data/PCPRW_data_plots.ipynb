```python
param_file='precipitation.csv'
df = pd.read_csv(param_file,delimiter='\t',skiprows=7)
#print df
days=Series(df.ix[:,0]).values
CARSNOW=Series(df.ix[:,1]).values
CRREL=Series(df.ix[:,2]).values
Helmers=Series(df.ix[:,3]).values
CPEAK=Series(df.ix[:,4]).values

plt.figure(1)
plt.subplot(221)
myplot=plt.plot(np.array(CARSNOW, dtype=float)); 
#plt.xlabel('Time (days)')
plt.ylabel('[mm per day] Precipitation')
plt.title('CARSNOW')
plt.setp(myplot, 'color', 'b', 'linewidth', 3.0)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='on',      # ticks along the bottom edge are off
    top='on',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.grid(); 

plt.subplot(222)
plt.plot(np.array(CRREL, dtype=float)); 
#plt.xlabel('Time (days)')
#plt.ylabel('[mm per day] Precipitation')
plt.title('CRREL')
plt.setp(myplot, 'color', 'b', 'linewidth', 3.0)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='on',      # ticks along the bottom edge are off
    top='on',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
plt.grid(); 

plt.subplot(223)
plt.plot(np.array(Helmers, dtype=float)); 
plt.xlabel('Time (days)')
plt.ylabel(' Precipitation')
plt.title('Helmers')
plt.setp(myplot, 'color', 'b', 'linewidth', 3.0)
plt.grid(); 

plt.subplot(224)
plt.plot(np.array(CPEAK, dtype=float)); 
plt.xlabel('Time (days)')
#plt.ylabel('Precipitation')
plt.title('CPEAK')
plt.setp(myplot, 'color', 'b', 'linewidth', 3.0)
plt.grid(); 

plt.show()
```

```python
param_file='streamflow.csv'
df1 = pd.read_csv(param_file,delimiter='\t',skiprows=0)
#print df1
days=Series(df1.ix[:,0]).values
C2=Series(df1.ix[:,1]).values
C3=Series(df1.ix[:,2]).values

plt.figure(2)
plt.subplot(121)
plt.plot(np.array(C2, dtype=float)); 
plt.xlabel('Time (days)')
plt.ylabel(' Discharge [ft3/s]')
plt.title('C2')
plt.setp(myplot, 'color', 'b', 'linewidth', 3.0)
plt.grid(); 

plt.subplot(122)
plt.plot(np.array(C3, dtype=float)); 
plt.xlabel('Time (days)')
#plt.ylabel('[mm per day] Precipitation')
plt.title('C3')
plt.setp(myplot, 'color', 'b', 'linewidth', 3.0)
plt.grid(); 
```
